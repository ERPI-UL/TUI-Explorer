```python
import re

pattern = r'\((\d{4})\)'

def extract_date(json_obj):
    # Extract the description field (or any other relevant text fields)
    # description = json_obj.get("reference", "")
    
    # Search for the date pattern in the description
    match = re.search(pattern, json_obj["reference"])
    
    if match:
        year = match.group(1)  # Capture the year inside parentheses
        json_obj["year"] = int(year)  # Add a new parameter 'year' to the JSON object
```

```python
def splitInput(json_obj):
    inputs = ['Non Linear', 'Linear', 'Rotational', 'Binary', 'Free']
    obj_input_array = []
    for input in inputs:
        if input in json_obj['input']:
            obj_input_array.append(input)
    json_obj["input"] = obj_input_array
```

```python
def splitOutput(json_obj):
    outputs = ['Audio', 'Haptic', 'Visual']
    obj_output_array = []
    for output in outputs:
        if output in json_obj['output']:
            obj_output_array.append(output)
    json_obj["output"] = obj_output_array
```

```python
def splitDelivTech(json_obj):
    delivTechnologies = ['Wearables', 'Projection', 'Handheld object', 'Overlay AR']
    obj_delivTech_array = []
    for delivTech in delivTechnologies:
        if delivTech in json_obj['delivery_technology']:
            obj_delivTech_array.append(delivTech)
    json_obj["delivery_technology"] = obj_delivTech_array
```

```python
def splitRecepTech(json_obj):
    recTechnologies = ['Computer Vision', 'Microcontrollers, Sensors, and Actuators', 'RFID']
    obj_recepTech_array = []
    for recepTech in recTechnologies:
        if recepTech in json_obj['reception_technology']:
            obj_recepTech_array.append(recepTech)
    json_obj["reception_technology"] = obj_recepTech_array
```

```python
def splitObject(json_obj):
    objects = ['Geometric', 'Controller', 'Display', 'Organic','Non shape retaining','Shape retaining']
    obj_array = []
    for obj in objects:
        if obj in json_obj['object']:
            obj_array.append(obj)
    json_obj["object"] = obj_array
```

```python
def addImage(json_obj):
    exp_id = json_obj['ID']
    image_name = exp_id.lower() + ".jpg"
    json_obj["img"] = image_name
```

```python
import csv
import json
import re
pattern = r'\((\d{4})\)'

def csv_to_json(csv_file_path, json_file_path):
    # Initialize an empty list to store the JSON records
    data = []
    
    # Open the CSV file for reading
    with open(csv_file_path, mode='r', encoding='utf-8') as csvfile:
        # Create a CSV reader object
        csvreader = csv.DictReader(csvfile, delimiter=';')
        
        # Iterate over each row in the CSV file
        for row in csvreader:
            # Each row is already a dictionary (key: value pairs)
            data.append(row)

    for obj in data:
        # match = re.search(pattern, obj["reference"])
        # if match:        
        #     year = match.group(1)  # Capture the year inside parentheses
        #     print(year)
        #     obj["year"] = year  # Add a new parameter 'year' to the JSON object
        # extract_date(obj)
        splitInput(obj)
        splitOutput(obj)
        splitDelivTech(obj)
        splitRecepTech(obj)
        splitObject(obj)
        addImage(obj)
        if obj["year"]:
           obj["year"] = int(obj["year"])
    
    # Open the JSON file for writing
    with open(json_file_path, mode='w', encoding='utf-8') as jsonfile:
        # Write the data to the JSON file in pretty-printed format
        json.dump(data, jsonfile, indent=4)
```

```python
csv_input ="SoA-TUI.csv"
json_output = "dataTUI-v2.json"
```

```python
csv_to_json(csv_input, json_output)
```
